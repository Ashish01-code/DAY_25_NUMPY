```python
#MINI CODE FOR NUMPY STATISTICS
import numpy as np
num=[]
n=int(input("ENTER THE LIMIT OF THE LIST "))
print("ENTER THE ELEMENTS")
for i in range(n):
    num.append(int(input()))
print("THE LIST IS ",num)
arr=np.array(num)
print("NumPy Array:", arr)
print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)
print("THE TYPE",type(arr))
print("MEAN",np.mean(arr))
print("MEDIAN",np.median(arr))
print("STANDARD DEVIATION",np.std(arr))
print("NORMLISED DATA: ",(arr - np.mean(arr)) / np.std(arr)) 
```

```python
import numpy as np

num1 = []
num2 = []

n = int(input("ENTER THE LIMIT OF FIRST ARRAY: "))
print("PLEASE ENTER ELEMENTS FOR 1ST ARRAY")
for i in range(n):
    num1.append(int(input()))

m = int(input("ENTER THE LIMIT OF SECOND ARRAY: "))
print("PLEASE ENTER THE ELEMENTS OF SECOND ARRAY")
for i in range(m):
    num2.append(int(input()))

arr1 = np.array(num1)
arr2 = np.array(num2)

if n == m:
    while True:
        print("\n--- MENU ---")
        print("1. ADD ARRAY")
        print("2. SUBTRACT ARRAY")
        print("3. MULTIPLY")
        print("4. POWER")
        print("5. REMAINDER")
        print("6. QUOTIENT & REMAINDER")
        print("7. ABSOLUTE (arr1)")
        print("0. EXIT")

        choice = int(input("ENTER THE CHOICE: "))

        if choice == 1:
            print("SUM:", np.add(arr1, arr2))
        elif choice == 2:
            print("DIFFERENCE:", np.subtract(arr1, arr2))
        elif choice == 3:
            print("MULTIPLY:", np.multiply(arr1, arr2))
        elif choice == 4:
            print("POWER:", np.power(arr1, arr2))
        elif choice == 5:
            print("REMAINDER:", np.remainder(arr1, arr2))
        elif choice == 6:
            q, r = np.divmod(arr1, arr2)
            print("QUOTIENT:", q)
            print("REMAINDER:", r)
        elif choice == 7:
            print("ABSOLUTE:", np.absolute(arr1))
        else:
            print("EXITING......")
            break
else:
    print("BOTH ARRAYS SHOULD HAVE SAME LENGTH")
```
